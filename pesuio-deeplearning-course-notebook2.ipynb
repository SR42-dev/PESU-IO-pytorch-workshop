```python
import torch as tc
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
X1 = np.random.random(N)*10 - 5
X2 = np.random.random(N)*20 - 6
X3 = np.random.random(N)*30 - 7

w1 = 0.5 # aim is to find w & b such that
w2 = 0.6
w3 = 0.7
b = 1

Y = w1*X1 + w2*X2 + w3*X3 - b + np.random.random(N) # X & noise components are randomized
```

```python
plt.scatter(X1,Y)
plt.scatter(X2,Y)
plt.scatter(X3,Y)
```

```python
model = tc.nn.Linear(3,1) 
```

```python
e_func = tc.nn.MSELoss()
optim = tc.optim.SGD(model.parameters(), lr = 0.001)
```

```python
X = np.dstack([X1,X2,X3]) # or np.array([X1, X2, X3])
X = X.reshape(N,3) 
Y = Y.reshape(N,1) 

# print(X)
# print(Y)

x = tc.from_numpy(X.astype(np.float32))
y = tc.from_numpy(Y.astype(np.float32))
```

```python
ep = 1000 

for e in range(ep) :
    optim.zero_grad() 
    
    output = model(x) 
    loss = e_func(output, y)
    print('Loss : ', loss.item())
    
    loss.backward() 
    optim.step() 
```
